## Time Resampling

```python
df_tsla.head(6)
```

##Business End Frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78

```python
df_tsla.resample(rule = 'A').max()
```

```python
df_tsla.resample(rule = 'QS').max()
```

```python
df_tsla.resample(rule = 'QS').max()['High'].plot()
```

```python
# Rule = A yearly max or min..

df_tsla.resample(rule = 'A').max()['Open'].plot()
```

```python

df_tsla.resample('10D').max()['Open'].plot()
```

```python
df_tsla.resample('30D').max()['Open'].plot()
```

```python
df_tsla['Open'].plot(figsize = (12,4))
```

```python
fig, ax = plt.subplots(3, sharex=True)
df_tsla['Open'].plot(figsize=(12,8), ax=ax[0], title="Original Data",
legend=False)

df_tsla.resample('10D').max()['Open'].plot(figsize=(12,8), ax=ax[1], 
title="Down-sampled to 10-day periods",legend=False)

df_tsla.resample('30D').max()['Open'].plot(figsize=(12,8), ax=ax[2],
title="Down-sampled to 30-day periods",legend=False)
```

```python
##plotting
df_tsla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
##plotting
df_tsla['High'].resample(rule='M').mean().plot(kind='bar',figsize = (20,6))
```

## Expanding command for CMA (Cumulative moving average)

```python
df_tsla['Open'].expanding().mean().plot(figsize = (16,8))
```

## Exponenetial Moving Average

```python
df_tsla['EMA_0.1'] = df_tsla['Open'].ewm(alpha = 0.1, adjust = False).mean()
```

```python
df_tsla[['Open', 'EMA_0.1']].plot(xlim=['2020-01-01','2021-09-01'],figsize = (16,8))
```

```python
df_tsla['EMA_0.3'] = df_tsla['Open'].ewm(alpha = 0.3, adjust = False).mean()
```

```python
df_tsla[['Open', 'EMA_0.1','EMA_0.3']].plot(xlim=['2020-01-01','2021-09-01'],figsize = (16,8))
```
